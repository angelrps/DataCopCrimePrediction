# Data Adquisition

```python
# Request the Dataset

import requests

data = requests.get('https://phl.carto.com:443/api/v2/sql?format=CSV&q=select * from phl.incidents_part1_part2')
```

```python
import csv

with open('/home/eduardo/Desktop/Proyecto/Philadelphia/crime_philadelphia_20062019.csv', 'w') as f:
    writer = csv.writer(f)
    reader = csv.reader(data.text.splitlines())

    for row in reader:
        writer.writerow(row)
```
