### 1. Loading Libraries <a id='1'></a>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from fbprophet import Prophet
from statsmodels.tools.eval_measures import rmse

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)

# To avoid seeing warnings
import warnings
warnings.filterwarnings('ignore')
```

### 2. Reading the Dataset <a id='2'></a>

```python
path = '../data/CSV/'
file = 'city_districts.csv'

df = pd.read_csv(path+file)
df.head()
```

```python
#################################################
#####  Prophet for each district
#################################################

# Distrcits
districts = [x for x in df.columns if x not in ['dispatch_date']]
#districts = ['city','dist_1'] # Testing

# Date Column to Datetime
df.rename(columns={'dispatch_date':'ds'},inplace=True)
# Make sure ds column is a pandas datetime object
df['ds'] = pd.to_datetime(df['ds'])

# Dataframe where we are going to save our result
prophet_df = pd.DataFrame()

for k,dist in enumerate(districts):
    new_df = pd.DataFrame()
    new_df = df[['ds',dist]]
    new_df.rename(columns={dist:'y'},inplace=True)
    # Train Test Split: Validation Dataset: 1 Year
    train = new_df.iloc[:(len(new_df)-12)]
    test = new_df.iloc[(len(new_df)-12):]
    
    ######## Prophet Model ########
    m = Prophet(seasonality_mode='multiplicative') # 'additive' Find out what is better for your model: RMSE
    m.fit(train)
    
    # In Stead of doing df.index.freq = 'MS' as we do in Statsmodel
    future = m.make_future_dataframe(periods=12,freq='M')
    preds = m.predict(future)
    
#     m.plot(preds)
#     plt.title(dist);
    
#    ######## Predictions ######## 
#     if dist == 'city':
#         prophet_df = preds[['ds','yhat']]
#         prophet_df['district'] = dist
#     else:
#         new_prophet_df = preds[['ds','yhat']]
#         new_prophet_df['district'] = dist
#         prophet_df = prophet_df.append(new_prophet_df)
        
     
    ######## Evaluating the model ########  
    
    # Alternative: from sklearn.metrics import mean_squared_error
    error = rmse(test['y'],preds['yhat'][-12:])
    # it should be interesting to compare this value with test['y'].mean()
    print(dist)
    print('Error: %s\nTest Mean: %s'%(error,test['y'].mean()))
    
    if k == 0:
        rmse_df = pd.DataFrame({'rmse':[error],'testMean':[test['y'].mean()],'district':[dist],'model':'prophet'})
    else:
        new_rmse_df = pd.DataFrame({'rmse':[error],'testMean':[test['y'].mean()],'district':[dist],'model':'prophet'})
        rmse_df = rmse_df.append(new_rmse_df)  
    
    ######## Forecast ######## 
    
    mf = Prophet(seasonality_mode='multiplicative')
    mf.fit(new_df)
    # In Stead of doing df.index.freq = 'MS' 
    future_f = mf.make_future_dataframe(periods=12,freq='M')

    forecast = m.predict(future_f)
    
    m.plot(forecast)
    plt.title(dist);
    
    if dist == 'city':
        forecast_df = pd.DataFrame({'values':forecast['yhat'].values,'district':dist,'type':'Time Series'},index=forecast['ds'])
        forecast_df.iloc[-12:]['type'] = 'Forecast'
        forecast_df = forecast_df.iloc[-12:]
    else:
        new_forecast_df = pd.DataFrame({'values':forecast['yhat'].values,'district':dist,'type':'Times Series'},index=forecast['ds'])
        new_forecast_df.iloc[-12:]['type'] = 'Forecast'
        new_forecast_df = new_forecast_df.iloc[-12:]
        forecast_df = forecast_df.append(new_forecast_df)
```

```python
# Actual Values DataFrame
df.set_index('ds',drop=True,inplace=True)

for col in df.columns.tolist():
    if col == 'city':
        actual = pd.DataFrame({'values':df[col],'district':col,'type':'Time Series'},index=df.index)
    else:
        new_actual = pd.DataFrame({'values':df[col],'district':col,'type':'Time Series'},index=df.index)
        actual = actual.append(new_actual)
```

```python
# Actual Values + Forecasted Values DataFrame
actual_forecast = actual.append(forecast_df)
actual_forecast['model'] = 'prophet'
```

```python
# Create DISTRICT ID
def district_id(x):
    if x == 'city':
        return 0
    else:
        return x[5:]
rmse_df['DISTRICT_'] = rmse_df['district'].apply(district_id)
actual_forecast['DISTRICT_'] = actual_forecast['district'].apply(district_id)
```
