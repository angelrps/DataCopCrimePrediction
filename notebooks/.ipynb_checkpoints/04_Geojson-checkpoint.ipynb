# Getting Started on Geospatial Analysis

```python
import geopandas as gpd
```

```python
path_geojson = '../data/geodata/Boundaries_District.geojson'
districts = gpd.read_file(path_geojson)
print(districts.head())
```

```python
districts.head()
```

```python
districts.plot(alpha=1,figsize=(10,10), cmap='winter');
```

```python
districts.plot(alpha=0.1,figsize=(10,10), cmap='winter');
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Setup figure and axis
f, ax = plt.subplots(1,figsize=(10,10))
# Plot layer of polygons on the axis
districts.plot(ax=ax)
# Remove axis frames
ax.set_axis_off()
# Display
plt.show()
```

```python
# Setup figure and axis
f, ax = plt.subplots(1,figsize=(10,10))
# Plot layer of polygons on the axis
districts.plot(linewidth=0.5, facecolor='green', edgecolor='blue', ax=ax,alpha=0.5)
# # Remove axis frames
# ax.set_axis_off()
# Display
plt.show()
```

```python
import pandas as pd
import numpy as np
from bokeh.io import show, output_notebook
from bokeh.models import GeoJSONDataSource, LinearColorMapper

import geopandas as gpd
from bokeh.palettes import Viridis6 as palette
from bokeh.plotting import figure
```

```python
output_notebook()


#obtain countries shapes
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
europe = (world.loc[world['continent'] == 'North America'])


geo_source = GeoJSONDataSource(geojson=europe.to_json())

p = figure(title="North America", x_range=(-30,60), y_range=(30,85))

p.patches('xs', 'ys', fill_alpha=0.7,
          fill_color={'field': 'gdp_md_est', 'transform': LinearColorMapper(palette=palette)},
          line_color='black', line_width=0.5, source=geo_source)

show(p)
```
