### 1. Loading Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

### 2. Loading Files

```python
sarima = pd.read_csv('../data/CSV/SarimaForecast_allDistricts_Tableau.csv',index_col=0)
prophet = pd.read_csv('../data/CSV/ProphetForecast_allDistricts_Tableau.csv',index_col=0)
rmse_sarima = pd.read_csv('../data/CSV/rmse_sarima.csv')
rmse_prophet = pd.read_csv('../data/CSV/rmse_prophet.csv')
# keras
# keras rmse
```
